### Missing Value Analysis: Churn Reason

The dataset contains 5,174 missing values in the `Churn Reason` column. Cross-tabulation with the target variable confirmed that all missing values belong to customers who did not churn, while all 1,869 customers who churned have a recorded churn reason.

Therefore, the missing values in `Churn Reason` are structurally expected rather than accidental data-quality issues. Customers who remain with the company naturally do not have a reason for churn.

The `Churn Reason` column will be retained for descriptive and business analysis of churned customers but excluded from the machine learning input features because it represents post-churn information and would introduce target leakage.

### Duplicate and Target Distribution Analysis

No duplicate records were found in the dataset, indicating that each row represents a unique customer record.

The target variable, `Churn Label`, contains 5,174 non-churned customers (73.46%) and 1,869 churned customers (26.54%). This indicates a moderate class imbalance, with non-churned customers forming the majority class.

Because the objective of the project is to identify customers at risk of leaving, model performance will not be evaluated using accuracy alone. Metrics such as precision, recall, F1-score, ROC-AUC, and PR-AUC will also be considered during model evaluation. Appropriate class-imbalance handling techniques will be evaluated during the modeling phase if required.
